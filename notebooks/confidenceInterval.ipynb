```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import bootstrap
```

```python
pennie_samples = np.array([2002, 1986, 2017, 1988, 2008, 1983, 2008, 1996, 2004, 2000, 
                           1994, 1995, 2015, 1978, 1974, 2015, 2016, 1996, 1983, 1971, 
                           1981, 1976, 1998, 2017, 1979, 1979, 1993, 2006, 1988, 1978, 
                           2013, 1976, 1979, 1985, 1985, 2015, 1962, 1999, 2015, 1990,
                           1992, 1997, 2018, 2015, 1997, 2017, 1982, 1988, 2006, 2017])
```

```python
plt.figure(1, figsize=(16,8))
_ = plt.hist(pennie_samples, bins=7)
```

```python
np.mean(pennie_samples)
```

```python
Replications = np.array([np.random.choice(pennie_samples, len(pennie_samples), replace = True) for _ in range(1000)])
```

```python
Replications[999]
```

```python
plt.figure(1, figsize=(16,8))
plt.hist(Replications[63], bins=7)
```

```python
arr = np.mean(Replications,axis=1)
```

```python
upper_bound = np.quantile(arr, 0.975)
lower_bound = np.quantile(arr, 0.025)
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
#ax=plt.figure(1, figsize=(16,8))
w=1
plt.hist(arr, bins=np.arange(min(arr), max(arr) + w, w))
_ = plt.axvline(x=lower_bound, ymax=0.95, linewidth=1.5, color="red")
_ = plt.axvline(x=upper_bound, ymax=0.95, linewidth=1.5, color="red")
_ = ax.set_xlabel('Resample sample mean')
_ = ax.set_ylabel('Count')
```

```python
np.mean(arr)
```

```python
print(lower_bound.round(2), upper_bound.round(2))
```

```python
#convert array to sequence
data = (pennie_samples,)

#calculate 95% bootstrapped confidence interval for median
bootstrap_ci = bootstrap(data, np.mean, confidence_level=0.95,
                         random_state=1, method='percentile')

#view 95% boostrapped confidence interval
print(bootstrap_ci.confidence_interval)
```
